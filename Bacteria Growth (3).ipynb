### Bacteria Growth <a name="bgrowth"></a>

Bacteria reproduce asexually by binary fission wherby each cell splits into 2 over a given time period. It is a unique characteristic of all living organisms and is affected by many various factors such as: pH, temperature, osmotic and hydrostatic pressure, moisture, nutritional factors. 

The time period that it takes for a species of bacteria to divide by binary fission is called its generation time and it is specific for different species of bacteria. Furthermore, generation times differ based on the growth medium that the bacteria is in. The factors listed above are examples of how a growth medium can be different. Every bacterial species has an optimum growth medium where their generation times are the fasted. 

**There are 4 main characteristics of the growth cycle of bacteria:**

**1. Lag phase:**  Immediately after inoculation of the cells into the growth medium, the population temporarily remains unchanged. The cells are preparing for binary fission by synthesising proteins, enzymes and RNA. Hence, they grow in volume/mass but they do not divide.

**2. Exponential phase:** In this phase the bacteria are regulary dividing and are growing by geometric progression. The time it takes for the bacteria to divide into 2 is individual to the bacterial species, since different species have different generation times. This growth phase will continue indefinetly until a limiting factor which prevents growth develops.

**3. Stationary phase:** If there is a limiting factor that prevents growth the bacterial population will stop growing. These factors could be a lack of nutrients, water, space or an accumulation of inihibitory metabolites. Hence the population size remains unchanged. 

**4. Death:** If the population reaches a stationary phase, a death phase will follow. In this stage, the number of viable cells will decrease exponentially.

# E.coli growth experiment <a name="E.coligrowth"></a>

Bacterial growth is time and temperature dependent process in which heat is released. This is because bacterial growth is a metabolic process which all exothermic and release heat. Hence, using a calorimeter, you can produce a thermogram of bacterial growth and determine the rate of growth of bacteria. The rate constant, k, quantifies this rate of reaction.

Following a microcalorimetric study of bacterial growth completed by Xie Chang-Li and others, the growth thermograms of E.coli have been determined. These thermogram curves reflect the changes in E.coli growth during the lag phase, exponential growth phase, stationary phase and decline phase of growth. The experiments are highly reproducible and can therefore be used as fingerprint in order to determine that E.coli is growing on a surface.

### Calculation of the bacterial multiplication rate constant, k <a name="multirateconst"></a>

Using bacteria population numbers and time, I was able to calculate the multiplication rate constant, $k$. 

If the bacterial number is $n_0$ at time $t_0$ and $n_t$ at time $t$, then:

$n_t = n_0 e^{k(t - t_0)}$

Where $k$ is the multiplication rate constant


When $t_0$ = 0:

$n_t = n_0 e^{k t}$


If the power output for each bacterium is $w$:

$n_{t} w = n_0 w e^{kt}$ 


We can then write $W_0 = n_0 w$  and  $W_t = n_t w$ which gives:

$W_t = W_0 e^{kt}$   or   $ln(W_t) = ln(W_0) + kt$


In the report it states that $lnW_0 = 0.04798$ so $W_0$ = 1.049.

### Calculation of Work Done <a name="workdone"></a>

Hence, the average value of the rate constant of multiplication for E.coli growth is 0.03968 $min^{-1}$

Therefore the work done by E.coli during multiplication, $W_t$ is as follows:

$W_t = W_0 e^{0.03968t}$      where $W_0$ is the power output of the bacteria present at t = 0 

As stated above, $lnW_0 = 0.04798$ so $W_0$ = 1.049. 

Hence, overall:

$W_t = 1.049e^{0.03968t}$



The data from this experiment can therefore be used to estimate accurately the generation time of E.coli. 

Using the equation $t = \frac{ln2}{k}$ with k = 0.03968 we get that t = 17.46 minutes

# Modelling  bacteria growth using differential equations <a name="diffequations"></a>

However there is a more accurate method for calculating k. Since bacterial growth is modelled by an ordinary differential equation, we can solve the ODE and get a value for k which is not rounded up/down and uses exact values.

Bacterial growth can be modelled by a differential equation since it depends on an inital condition (intitial number of cells) and a constantly changing rate that is linked with the number of cells. 

In this equation we let the number of bacteria in a population be denoted by P. Since the number of bacteria varies with time we know P is a function of time, P(t). Now, since we know that bacteria divide by binary fission, we realise that the rate of growth of the population of bacteria is proportional to the size of the population. 

Hence, $ \frac{dP}{dt} \propto P(t) $ 

Which is written in a differential equation as such:

$\frac{dP}{dt} = k P(t) $   

Where k is a constant of proportionality which represents the rate of growth and k cannot equal 0. 

This equation is correct as a model for all bacteria if there are no limiting factors of growth such as moisture, heat, nutrients. 

Solving this differential equation gives you $ P(t) = e^{kt}$

We can then solve this for k. We know that during binary fission a bacterium splits into 2 but the time it takes for a bacterium to do this depends on its generation times. 

If we let the generation time of the bacteria to equal t, we know that $P(t) = 2(0)$ since in the time from 0 to its generation time the population doubled. Hence we know the following:

$e^{kt} = 2P(0) = 2e^{0} = 2$ 

So, $kt = ln2$

So, $k = \frac{1}{t} ln2 min^{-1}$

We can use the inverse of its generation times multipled by $ln2$ to determine the rate of growth so that if the generation time is longer, the rate of growth is slower. 

### Errors with bacterial growth modelling <a name="modelerror"></a>

Since, the mean of k is 0.03968 min ^ -1 has a standard deviation of 0.00257 min ^ -1, using the mean leads to errors arising. 

By using 0.03968  ±  0.00257 min ^ -1 I have calculated the lower estimate of k, 0.03711 min ^ -1 and the upper estimate of k, 0.04225 min ^ -1. 

Hence, a ranged absolute error for k can be calculated using $ | \frac{1}{t} ln2 min^{-1} - k | $ 

The absolute error of using $0.03968 min^{-1}$ instead of $\frac{1}{t} ln2 min^{-1}$ is $0.00109336356 min^{-1}$

# Effect of temperature on bacterial growth experiment <a name="tempgrowth"></a>

Temperature is one of the main factors controlling bacterial growth. Bacteria are infact classified based on their temperature range at which it can grow at. Organisms categorized as mesophiles are adapted to moderate temperatures, with optimal growth temperatures ranging from room temperature (about 20 °C) to about 45 °C. Organisms called psychrotrophs prefer cooler environments, from a high temperature of 25 °C to refrigeration temperature about 4 °C. They are found in many natural environments in temperate climates. They are also responsible for the spoilage of refrigerated food. Organisms that grow at optimum temperatures of 50 °C to a maximum of 80 °C are called thermophiles. They do not multiply at room temperature. Thermophiles are widely distributed in hot springs, geothermal soils. 

E.coli, Salmonella and Lactobacillus are examples of mesophiles whereas Arthrobacter and Psychrobacter are members of the psychrotrophic bacterial class. Examples of thermophiles include Thermus aquaticus and Geobacillus.

# Conclusion <a name="conclusion"></a>


From the following experiment I can conclude that temperature does have an effect on the growth of bacteria. It is clear to see that with an increase in temperature, there is an increase of microbial growth up to an optimum temperature. After this temperature is reached, microbial growth decreases rapidly. This makes sense in line with the structure of bacteria that we know and the process of binary fission. 

During the lag phase, enzymes that are needed for binary fission are synthesised. Collisions between all molecules in a bacterial cell increase as temperature increases. This is due to the increase in velocity and kinetic energy that follows temperature increases. With faster velocities, there will be less time between collisions. This results in more collisions between enzymes and substrates so the rate of binary fission increases. 

Enzymes have an optimum temperature at which they work for and any increased temperature causes enzyme denaturation. This means that the structure of the enzyme is changed by the high energy. In particular, the active site, which defines the enzymes function. Hence, when a bacterium is grown in a more than optimum temperature, the enzymes needed for binary fission denature so the bacteria cannot divide or survive. 

If the bacteria was left for longer, exponential growth would likely slow and stop since there would be a limiting factor of space. If you see from the above graph at warmer temperatures, almost all of the petri dish was covered in bacteria. Hence, after 100% of the petri dish was covered, the bacteria would be competing for space, nutrients and moisture and some bacteria would die. After this happens, the stationary phase of growth followed by death would occur. 

For most bacteria, that can survive in humans, and hence are those that pose a threat to human life, grow at an optimum of 37 degrees celsius. This is the same for the bacteria in this experiment: Salmonella enterica, Salmonella bongori, Staphylococcus aureus and Listeria monocytogenes. Hence the results from this experiment can be used as a fingerprint for other mesophile bacterial populations that may cause threat to human life.
